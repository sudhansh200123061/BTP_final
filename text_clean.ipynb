```python
import glob
import re
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import emoji
import os

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

Myfiles = {}
filname = {}
path1 = 'D:/7thsem/btp/final/20news-18828/'
path2 = 'D:/7thsem/btp/final/cleaned/'

ps = PorterStemmer()
stop_words = set(stopwords.words('english'))
stop_words.add('aa')
stop_words.add('aaaahhh')
stop_words.add('aah')
stop_words.add('aam')
stop_words.add('aap')
dir_list = os.listdir(path1)
```

```python
def remove_lines_with_pattern(text):
    pattern = r'\S*@\S*'
    digit = r'\d+'
    lines = text.split('\n')
    new_lines = []
    lemmatizer = WordNetLemmatizer()
    for line in lines:

        if not re.search(pattern, line):
            new_line = re.sub(digit, '', line)
            new_line = emoji.demojize(new_line)
            new_line = re.sub(r'\s\S*\www\S*\s',' ',new_line)
         
            new_line = re.sub(r'[^a-zA-Z]', ' ', new_line)
            
            new_line = re.sub(r'\bSubject\b', '', new_line)
            new_line = new_line.lower()
            word_list = nltk.word_tokenize(new_line)
            filtered_words = [word for word in word_list if not word in stop_words]
            stemmed_words = [ps.stem(word) for word in filtered_words]
            filtered_words = [word for word in stemmed_words if not word in stop_words]
            new_line = ' '.join([lemmatizer.lemmatize(w) for w in filtered_words])
            new_line = re.sub(r'\bre\b', '', new_line)
            new_lines.append(new_line)
    return ' '.join(new_lines)
   

def text_clean(data,j,fol):
    data = remove_lines_with_pattern(data) 
    os.makedirs(os.path.dirname(path2+fol+"/"), exist_ok=True)
    open(path2+fol+"/"+j,'w').write(data)
    # print(data)
    

for fol in dir_list:
    Myfiles[fol] = [I for I in glob.glob(path1+fol+"/*.txt")]

for fol in dir_list:
    filname[fol] = []
    for i in Myfiles[fol]:
        name = i.split('/')
        name = name[len(name)-1].split('\\')
        filname[fol].append(name[1])
        # print(name[1])


for fol in dir_list:
    for i,j in zip(Myfiles[fol],filname[fol]):
        with open(i,'rb') as fin:
            data = fin.read().decode('utf-8', 'ignore')
            # print(j)
            text_clean(data,j,fol)
   
# for i,j in zip(Myfiles,names):
#     with open(i,'rb') as fin:
#         data = fin.read().decode('utf-8', 'ignore')
#         # print(data)
#         print(j)
#         text_clean(data)
        
        


```
